# 04 - Phân tích Grad-CAM

Hiển thị các mẫu Grad-CAM đã sinh. Chạy trước: `python -m src.explainability.generate_gradcam_samples --all-models`

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from src.utils.config import load_config

config = load_config()
model_name = 'resnet18'
gradcam_dir = config.output_root / 'figures' / 'gradcam' / model_name
samples = sorted(gradcam_dir.glob('*.png'))
print('Số mẫu:', len(samples))
```

```python
for path in samples[:6]:
    img = mpimg.imread(path)
    plt.figure(figsize=(12, 4))
    plt.imshow(img)
    plt.axis('off')
    plt.title(path.name)
    plt.show()
```
